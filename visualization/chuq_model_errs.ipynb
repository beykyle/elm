```python
import jitr
import json
import numpy as np
```

```python
from jitr.data import data_dir
```

```python
for f in (data_dir / "CHUQFederal/0").iterdir():
    if f.is_file():
        print(f)
```

```python
ecs_p = np.zeros(jitr.optical_potentials.chuq.NUM_POSTERIOR_SAMPLES)
ecs_n = np.zeros(jitr.optical_potentials.chuq.NUM_POSTERIOR_SAMPLES)
for i in range(jitr.optical_potentials.chuq.NUM_POSTERIOR_SAMPLES):
    with open(data_dir / f"CHUQFederal/{i}/modelErrors.json", "rb") as f:
        d = json.load(f)
        ecs_p[i] = d["ECS_p"]
        ecs_n[i] = d["ECS_n"]
```

```python
from matplotlib import pyplot as plt
```

```python
plt.hist(ecs_p)
```

```python
plt.hist(ecs_n)
```
